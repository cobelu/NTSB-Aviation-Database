```python
%config IPCompleter.use_jedi = False
import pandas as pd
```

```python
engine_df = pd.read_csv('../data/engines.csv')
events_df = pd.read_csv('../data/events.csv')
findings_df = pd.read_csv('../data/Findings.csv')
aircraft_df = pd.read_csv('../data/aircraft.csv')
injury_df = pd.read_csv('../data/injury.csv')
```

```python
corvair = 'CORVAIR'
eng_mfgr = 'eng_mfgr'
eng_model = 'eng_model'
corvair_mfr = engine_df[eng_mfgr].str.contains(corvair, False)
corvair_model = engine_df[eng_model].str.contains(corvair, False)
# corvair_mfr.head()
engine_df = engine_df[corvair_mfr | corvair_model]
```

```python
df = pd.merge(events_df, engine_df, on='ev_id')
df = pd.merge(df, findings_df, on='ev_id')
df = pd.merge(df, aircraft_df, on='ev_id')
# df = pd.merge(df, injury_df, on='ev_id')
```
